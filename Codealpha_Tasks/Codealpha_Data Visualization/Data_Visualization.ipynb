                         **Data Visualization**

```python
#Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset (Excel or CSV)
df = pd.read_csv('StudentPerformance.csv')
```

```python
#checking The Data
print(df.head())
```

         ***Data Visualization Codes***

**1. Bar Chart – Average Performance by Extracurricular Activity**

*Shows whether extracurricular activities affect performance*

```python
plt.figure()
sns.barplot(x='Extracurricular Activities', y='Performance Index', data=df,
            color="y")
plt.title('Average Performance by Extracurricular Activity')
plt.show()
```

  **2. Line Chart – Performance Index Trend**

*Shows overall performance trend*

```python
plt.figure()
plt.plot(df['Performance Index'])
plt.title('Performance Index Trend')
plt.xlabel('Student Index')
plt.ylabel('Performance Index')
plt.show()
```

**3. Histogram – Distribution of Performance Index**

*Shows how scores are distributed*

```python
plt.figure()
sns.histplot(df['Performance Index'], kde=True)
plt.title('Distribution of Performance Index')
plt.show()
```

**4. Scatter Plot – Sleep Hours vs Performance**

*Shows relationship between sleep and performance*

```python
plt.figure()
sns.scatterplot(x='Sleep Hours', y='Performance Index', data=df,color="g")
plt.title('Sleep Hours vs Performance Index')
plt.show()
```

**5. Scatter Plot – Sample Questions vs Performance**

  *Shows practice impact*

```python
plt.figure()
sns.scatterplot(x='Sample Question Papers Practiced', y='Performance Index',
                data=df,color="gold")
plt.title('Practice vs Performance Index')
plt.show()
```

**6. Box Plot – Previous Score vs Performance**

*Detects spread and outliers*

```python
plt.figure()
sns.boxplot(y='Previous Scores', data=df,color="brown")
plt.title('Previous Score Distribution')
plt.show()
```

**7. Heatmap – Correlation Between Numerical Features**

*Shows relationships between numerical columns*

```python
plt.figure()
corr = df[['Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced',
           'Performance Index']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm',color="r")
plt.title('Correlation Heatmap')
plt.show()
```
